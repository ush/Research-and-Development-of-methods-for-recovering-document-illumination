# Imports

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.gridspec as gridspec
from scipy import stats
```

# Loading data

```python
def load_data(name, metric):
    name = name + ".txt"
    data = pd.read_csv(name, sep=", ", header=None)
    data.columns = ["id", metric]
    return data
```

**PSNR**

```python
classic_SF_PSNR = load_data("classic_SF_PSNR", "PSNR")
cropped_SF_PSNR = load_data("cropped_SF_PSNR", "PSNR")
mixed_SF_PSNR = load_data("mixed_SF_PSNR", "PSNR")

classic_SG_PSNR = load_data("classic_SG_PSNR", "PSNR")
cropped_SG_PSNR = load_data("cropped_SG_PSNR", "PSNR")
mixed_SG_PSNR = load_data("mixed_SG_PSNR", "PSNR")

classic_STCGAN_PSNR = load_data("classic_STCGAN_PSNR", "PSNR")
cropped_STCGAN_PSNR = load_data("cropped_STCGAN_PSNR", "PSNR")
mixed_STCGAN_PSNR = load_data("mixed_STCGAN_PSNR", "PSNR")
```

**RMSE**

```python
classic_SF_RMSE = load_data("classic_SF_RMSE", "RMSE")
cropped_SF_RMSE = load_data("cropped_SF_RMSE", "RMSE")
mixed_SF_RMSE = load_data("mixed_SF_RMSE", "RMSE")

classic_SG_RMSE = load_data("classic_SG_RMSE", "RMSE")
cropped_SG_RMSE = load_data("cropped_SG_RMSE", "RMSE")
mixed_SG_RMSE = load_data("mixed_SG_RMSE", "RMSE")

classic_STCGAN_RMSE = load_data("classic_STCGAN_RMSE", "RMSE")
cropped_STCGAN_RMSE = load_data("cropped_STCGAN_RMSE", "RMSE")
mixed_STCGAN_RMSE = load_data("mixed_STCGAN_RMSE", "RMSE")
```

**WER**

```python
classic_SF_WER = load_data("classic_SF_WER", "WER")
cropped_SF_WER = load_data("cropped_SF_WER", "WER")
mixed_SF_WER = load_data("mixed_SF_WER", "WER")

classic_SG_WER = load_data("classic_SG_WER", "WER")
cropped_SG_WER = load_data("cropped_SG_WER", "WER")
mixed_SG_WER = load_data("mixed_SG_WER", "WER")

classic_STCGAN_WER = load_data("classic_STCGAN_WER", "WER")
cropped_STCGAN_WER = load_data("cropped_STCGAN_WER", "WER")
mixed_STCGAN_WER = load_data("mixed_STCGAN_WER", "WER")
```

**SF**

```python
classic_SF = pd.concat([classic_SF_PSNR['id'], classic_SF_PSNR['PSNR'], classic_SF_WER['WER'], 
                           classic_SF_RMSE['RMSE']], axis=1)

cropped_SF = pd.concat([cropped_SF_PSNR['id'], cropped_SF_PSNR['PSNR'], cropped_SF_WER['WER'], 
                           cropped_SF_RMSE['RMSE']], axis=1)

mixed_SF = pd.concat([mixed_SF_PSNR['id'], mixed_SF_PSNR['PSNR'], mixed_SF_WER['WER'], 
                           mixed_SF_RMSE['RMSE']], axis=1)
```

**SG**

```python
classic_SG = pd.concat([classic_SG_PSNR['id'], classic_SG_PSNR['PSNR'], classic_SG_WER['WER'], 
                           classic_SG_RMSE['RMSE']], axis=1)

cropped_SG = pd.concat([cropped_SG_PSNR['id'], cropped_SG_PSNR['PSNR'], cropped_SG_WER['WER'], 
                           cropped_SG_RMSE['RMSE']], axis=1)

mixed_SG = pd.concat([mixed_SG_PSNR['id'], mixed_SG_PSNR['PSNR'], mixed_SG_WER['WER'], 
                           mixed_SG_RMSE['RMSE']], axis=1)
```

**STCGAN**

```python
classic_STCGAN = pd.concat([classic_STCGAN_PSNR['id'], classic_STCGAN_PSNR['PSNR'], classic_STCGAN_WER['WER'], 
                           classic_STCGAN_RMSE['RMSE']], axis=1)

cropped_STCGAN = pd.concat([cropped_STCGAN_PSNR['id'], cropped_STCGAN_PSNR['PSNR'], cropped_STCGAN_WER['WER'], 
                           cropped_STCGAN_RMSE['RMSE']], axis=1)

mixed_STCGAN = pd.concat([mixed_STCGAN_PSNR['id'], mixed_STCGAN_PSNR['PSNR'], mixed_STCGAN_WER['WER'], 
                           mixed_STCGAN_RMSE['RMSE']], axis=1)
```

# Friedman



```python
classic_Friedman_WER = stats.friedmanchisquare(classic_STCGAN['WER'], classic_SG['WER'], classic_SF['WER'])
cropped_Friedman_WER = stats.friedmanchisquare(cropped_STCGAN['WER'], cropped_SG['WER'], cropped_SF['WER'])
mixed_Friedman_WER = stats.friedmanchisquare(mixed_STCGAN['WER'], mixed_SG['WER'], mixed_SF['WER'])

print("WER")
print("Friedman test statistic for classic WER: ", classic_Friedman_WER.statistic, " p-value: ", "{:.5f}".format(classic_Friedman_WER.pvalue))
print("Friedman test statistic for cropped WER: ", cropped_Friedman_WER.statistic, " p-value: ", "{:.5f}".format(cropped_Friedman_WER.pvalue))
print("Friedman test statistic for mixed WER: ", mixed_Friedman_WER.statistic, " p-value: ", "{:.5f}".format(mixed_Friedman_WER.pvalue))
```

```python
classic_Friedman_RMSE = stats.friedmanchisquare(classic_STCGAN['RMSE'], classic_SG['RMSE'], classic_SF['RMSE'])
cropped_Friedman_RMSE = stats.friedmanchisquare(cropped_STCGAN['RMSE'], cropped_SG['RMSE'], cropped_SF['RMSE'])
mixed_Friedman_RMSE = stats.friedmanchisquare(mixed_STCGAN['RMSE'], mixed_SG['RMSE'], mixed_SF['RMSE'])

print("RMSE")
print("Friedman test statistic for classic RMSE: ", classic_Friedman_RMSE.statistic, " p-value: ", "{:.5f}".format(classic_Friedman_RMSE.pvalue))
print("Friedman test statistic for cropped RMSE: ", cropped_Friedman_RMSE.statistic, " p-value: ", "{:.5f}".format(cropped_Friedman_RMSE.pvalue))
print("Friedman test statistic for mixed RMSE: ", mixed_Friedman_RMSE.statistic, " p-value: ", "{:.5f}".format(mixed_Friedman_RMSE.pvalue))
```

```python
classic_Friedman_PSNR = stats.friedmanchisquare(classic_STCGAN['PSNR'], classic_SG['PSNR'], classic_SF['PSNR'])
cropped_Friedman_PSNR = stats.friedmanchisquare(cropped_STCGAN['PSNR'], cropped_SG['PSNR'], cropped_SF['PSNR'])
mixed_Friedman_PSNR = stats.friedmanchisquare(mixed_STCGAN['PSNR'], mixed_SG['PSNR'], mixed_SF['PSNR'])

print("PSNR")
print("Friedman test statistic for classic PSNR: ", classic_Friedman_PSNR.statistic, " p-value: ", "{:.5f}".format(classic_Friedman_PSNR.pvalue))
print("Friedman test statistic for cropped PSNR: ", cropped_Friedman_PSNR.statistic, " p-value: ", "{:.5f}".format(cropped_Friedman_PSNR.pvalue))
print("Friedman test statistic for mixed PSNR: ", mixed_Friedman_PSNR.statistic, " p-value: ", "{:.5f}".format(mixed_Friedman_PSNR.pvalue))
```
